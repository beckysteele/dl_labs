```python
import descarteslabs as dl
from descarteslabs.catalog import Product, properties as p
from descarteslabs.catalog import BandType
import numpy as np
```

```python
# Create polygon definition for GEOjson object

geometry = {
    'type': 'Polygon',
    'coordinates': [[
        [-93.52300099792355, 41.241436141055345],
        [-93.7138666, 40.703737],
        [-94.37053769704536, 40.83098709945576],
        [-94.2036617, 41.3717716],
        [-93.52300099792355, 41.241436141055345]
    ]]
}
# aoi_geometry
```

```python
# Create a scene and ctx is with the aoi_geometry polygon, using the product id I have access to

landsat8_collection = Product.get("landsat:LC08:PRE:TOAR")
# print(dir(landsat8_collection))
# print(type(landsat8_collection))

scenes, ctx = dl.scenes.search(
    geometry,
    products=landsat8_collection.id
)
scenes
# ctx
```

```python
# Create an N-dimensional array (stack) of RGB and cloud scenes 

for band in landsat8_collection.bands().filter(p.type == BandType.CLASS).sort("name"):
    print(band)

rgb_ndarray_stack = scenes.stack(
    "red green blue",
    ctx.assign(resolution=100)
)

cloud_ndarray_stack = scenes.stack(
    "qa_cloud",
    ctx.assign(resolution=100)
)
```

```python
# Creates a masked array using the median of that stack (Median Composite)
rgb_composite = np.ma.median(rgb_ndarray_stack, axis=0)

cloud_composite = np.ma.median(cloud_ndarray_stack, axis=0)

final_composite = rgb_composite - cloud_composite
```

```python
# Display median composite
dl.scenes.display(rgb_composite, title="RGB Median composite of Polygon", size=6)
dl.scenes.display(cloud_composite, title="Cloud Median composite of Polygon", size=6)
dl.scenes.display(final_composite, title="Bottom of Atmosphere ")
```
